```python
# 1 
# Import necessary libraries
from matplotlib import pyplot as plt
import pandas as pd

# load rankings data
wood = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
# load rankings data
steel= pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')

#print(wood[wood['Name'] == 'Boulder Dash'])


# 2
# Create a function to plot rankings over time for 1 roller coaster
def rank_time(name,park):
  dfwood = wood[(wood['Name']== name) & (wood['Park'] == park)]
  plt.plot(dfwood['Year of Rank'],dfwood['Rank'], marker = 'o')
  ax = plt.subplot()
  ax.set_yticks(dfwood.Rank)
  plt.title('El Toro Rankings')
  plt.ylabel('Ranking')
  plt.show()
  plt.clf()

put= rank_time('El Toro','Six Flags Great Adventure')
```

# Function Plots

```python
# 3
# Create a plot of El Toro ranking over time
def rank2_time(name1,name2,park1,park2):
  dfwood1 = wood[(wood['Name'] == name1) & (wood['Park'] == park1)]
  dfwood2 = wood[(wood['Name'] == name2) & (wood['Park'] == park2)]
# Create a plot of El Toro and Boulder dash hurricanes
  plt.plot(dfwood1['Year of Rank'],dfwood1.Rank, marker = 'o')
  plt.plot(dfwood2['Year of Rank'],dfwood2.Rank, marker = 's')
  plt.title('El Toro and Boulder Dash Rankings')
  plt.ylabel('Ranking')
  ay = plt.subplot()
  ay.set_yticks([1,2,3,4])
  ay.invert_yaxis()
  plt.legend([name1,name2], loc = 1)
  plt.show()
rank2_time('El Toro','Boulder Dash','Six Flags Great Adventure','Lake Compounce')
```

 ### top n rankings over time

```python
# 4
# Create a function to plot top n rankings over time
def top_rankings(df,n):
  top = df[(df['Rank'] <= n)]
# Create a plot of top n rankings over time
  
  for coaster in set(top['Name']):
    fig, ax = plt.subplots(figsize = (10,10))
    coaster_rankings = top[top['Name'] == coaster]
    ax.plot(coaster_rankings['Year of Rank'],coaster_rankings['Rank'],label = coaster,marker = 's')
    ax.set_yticks([i for i in range(1,6)])
    ax.invert_yaxis()
    ax.set_xticks(coaster_rankings['Year of Rank'])
    plt.title('Top 10 Rankings')
    plt.xlabel('Year')
    plt.ylabel('Ranking')
    plt.legend(loc = 4)
    plt.show()
  

top_rankings(wood, 5)
plt.clf()
```

```python
# 5
# load roller coaster data
roller_coaster = pd.read_csv('roller_coasters.csv')
print(roller_coaster.head())
# 6
# Create a function to plot histogram of column values
def roller_hist(df,col):
  plt.hist(df[col])
  plt.legend([col])
  plt.xlabel(col)
  plt.ylabel('Number of Roller Coasters')
  plt.show()
roller_hist(roller_coaster,'speed')
plt.clf()

# Create histogram of roller coaster speed
roller_hist(roller_coaster,'speed')

# Create histogram of roller coaster length
roller_hist(roller_coaster,'length')

# Create histogram of roller coaster number of inversions
roller_hist(roller_coaster, 'num_inversions')

# Create a function to plot histogram of height values
def roller_height(df,col):
  heights = df[df[col] <= 140]
  plt.hist(df[col].dropna())
  plt.legend([col])
  plt.xlabel(col)
  plt.ylabel('Number of Roller Coasters')
  plt.show()
  plt.clf()

# Create a histogram of roller coaster height
roller_height(roller_coaster,'height')
```

```python
# 7
# Create a function to plot inversions by coaster at park
def roll_inv(df, park):
    park_df = df[df['park'] == park]
    roller_coaster = park_df['name']
    inversions = park_df['num_inversions']
    plt.figure(figsize = (20, 15))
    ax = plt.subplot()
    ay = plt.subplot()
    plt.bar(range(len(roller_coaster)), inversions)
    ax.set_xticks(range(len(roller_coaster)))
    ax.set_xticklabels(roller_coaster)
    plt.xticks(rotation=45)
    plt.legend([park])
    plt.xlabel('Roller Coaster')
    plt.ylabel('Number of Inversions')
    plt.show()

# Create barplot of inversions by roller coasters
roll_inv(roller_coaster,'Walibi Belgium')
```

```python
# 8
# Create a function to plot a pie chart of status.operating
def roll_pie(df):
  open_df = df[df['status'] == 'status.operating']
  closed_df = df[df['status'] == 'status.closed.definitely']
  count = [len(open_df), len(closed_df)]
  labels = ['Opened', 'Closed']
  plt.figure(figsize = (5,5))
  plt.pie(count, autopct='%0.1f%%', labels = labels)
  plt.title('Operating State of Roller Coasters')
  plt.legend(labels)
  plt.axis('Equal')
  plt.legend(labels)
  plt.show()
    
# Create pie chart of roller coasters
roll_pie(roller_coaster)
```

```python
# 9
# Create a function to plot scatter of any two columns
def roll_scatter(df,col1,col2):
  
  x = range(len(df))
  plt.figure(figsize =(6,6))
  plt.scatter(x,df[col1], alpha =0.5)
  plt.scatter(x,df[col2], alpha =0.5)
  ax= plt.subplot()
  plt.title('Scatter Plot')
  plt.legend([col1,col2])
  ax.set_xlabel('Variables')
  ax.set_ylabel('Roller Coasters')
  plt.ylim(0,200)
  plt.show()
  plt.clf()
roll_scatter(roller_coaster,'height', 'num_inversions')

# Create a function to plot scatter of speed vs height
def speed_height_scatter(df,col1,col2):
  x = range(len(df))
  plt.figure(figsize =(6,6))
  plt.scatter(x,df[col1], alpha =0.5)
  plt.scatter(x,df[col2], alpha =0.5)
  ax= plt.subplot()
  plt.title('Scatter Plot')
  plt.legend([col1,col2])
  ax.set_xlabel('Variables')
  ax.set_ylabel('Roller Coasters')
  plt.ylim(0,200)
  plt.show()
  plt.clf()
speed_height_scatter(roller_coaster,'speed', 'height')

# Create a scatter plot of roller coaster height by speed
plt.scatter(roller_coaster['speed'],roller_coaster['height'])
plt.xlabel('speed')
plt.ylabel('height')
plt.legend(['Height by Speed'])
plt.ylim(0,200)
plt.show()
```
